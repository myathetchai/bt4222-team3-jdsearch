## Candidate Encoding

Encode categorical IDs in product_df
- assign integer indices to each unique ID (starting from 1, 0 = padding/UNK)
- the mappings are stored in a dictionary e.g. wid2idx = {84294794: 12, ...}

```python
# Create a copy (just in case)
prod = product_df.copy()

def encode_col(df, colname):
    """
    Map raw IDs in df[colname] to contiguous indices:
    0 reserved for PAD/UNK.
    Returns: index_series, rawID -> index dict
    """
    # Ensure we treat as integers where possible
    df[colname] = df[colname].astype("Int64")

    uniques = df[colname].dropna().unique()

    # Map each unique non-null raw ID to 1..N
    id2idx = {int(v): i+1 for i, v in enumerate(uniques)}

    # For each value in the column, try to find the new ID, if not 0
    def mapper(x):
        if pd.isna(x):
            return 0
        return id2idx.get(int(x), 0)  # unseen -> 0 (UNK)

    idx_series = df[colname].map(mapper).astype(int)
    return idx_series, id2idx

# 1) wid ("primary id")
prod["wid_idx"], wid2idx = encode_col(prod, "wid")

# 2) brand, shop, category levels
prod["brand_idx"], brand2idx = encode_col(prod, "brand_id")
prod["shop_idx"],  shop2idx  = encode_col(prod, "shop_id")

prod["cate1_idx"], cate1_2idx = encode_col(prod, "cate_id_1")
prod["cate2_idx"], cate2_2idx = encode_col(prod, "cate_id_2")
prod["cate3_idx"], cate3_2idx = encode_col(prod, "cate_id_3")
prod["cate4_idx"], cate4_2idx = encode_col(prod, "cate_id_4")
```

Align rows + build lookup tensors
- Given wid_idx, we want to quickly know its brand/shop/cate indices using tensors
- A lookup tensor is just a 1-D array where: the position = key (wid_idx) and the value = mapped attribute (brand_idx)
- brand_lookup[wid_idx] -> brand_idx, shop_lookup[wid_idx] -> shop_idx, etc.
- Row 0 is all zeros (PAD/UNK)
| wid_idx | brand_idx | shop_idx | cate1_idx | cate2_idx | cate3_idx | cate4_idx |               |
| ------- | --------- | -------- | --------- | --------- | --------- | --------- | ------------- |
| 0       | 0         | 0        | 0         | 0         | 0         | 0         |               |
| 1       | 11        | 3        | 7         | 2         | 9         | 1         |               |
| 2       | 4         | 8        | 2         | 7         | 5         | 4         |               |
| **12**  | **5**     | **9**    | **7**     | **2**     | **6**     | **3**     | ← our product |
| ...     | ...       | ...      | ...       | ...       | ...       | ...       |               |


```python
# Sort the dataframe by wid index (row 1 in the dataframe corresponds to wid_idx = 1, row 2 → wid_idx = 2, etc.)
prod = prod.sort_values("wid_idx").reset_index(drop=True)

# Vocabulary sizes / No. of unique IDs (+1 cause we reserved index 0 as a special “padding / unknown” ID)
num_wid   = int(prod["wid_idx"].max()) + 1
num_brand = int(prod["brand_idx"].max()) + 1
num_shop  = int(prod["shop_idx"].max()) + 1
num_c1    = int(prod["cate1_idx"].max()) + 1
num_c2    = int(prod["cate2_idx"].max()) + 1
num_c3    = int(prod["cate3_idx"].max()) + 1
num_c4    = int(prod["cate4_idx"].max()) + 1

# Initialise lookup tables; index = wid_idx; value = corresponding feature idx
brand_lookup = torch.zeros(num_wid, dtype=torch.long)
shop_lookup  = torch.zeros(num_wid, dtype=torch.long)
c1_lookup    = torch.zeros(num_wid, dtype=torch.long)
c2_lookup    = torch.zeros(num_wid, dtype=torch.long)
c3_lookup    = torch.zeros(num_wid, dtype=torch.long)
c4_lookup    = torch.zeros(num_wid, dtype=torch.long)

for _, row in prod.iterrows():
    w = int(row["wid_idx"])
    if w == 0:
        continue  # padding / unk
    brand_lookup[w] = int(row["brand_idx"])
    shop_lookup[w]  = int(row["shop_idx"])
    c1_lookup[w]    = int(row["cate1_idx"])
    c2_lookup[w]    = int(row["cate2_idx"])
    c3_lookup[w]    = int(row["cate3_idx"])
    c4_lookup[w]    = int(row["cate4_idx"])
```

Map candidate_wid_list → wid_idx in user_behaviour_df
- [36098226, 82122191, 34768767, 71429327, 87635] → [15, 7, 12, 23, 19]

```python
# Create a copy (just in case)
ub = user_behaviour_df.copy()

def map_wids_to_idx(wid_list):
    safe_indices = []
    for w in wid_list:
        # Skip if missing or invalid
        if pd.isna(w) or w == 'nan' or w is None:
            safe_indices.append(0)  # 0 = padding/unknown
            continue
        try:
            wid_int = int(w)
        except (ValueError, TypeError):
            wid_int = 0
        safe_indices.append(wid2idx.get(wid_int, 0))
    return safe_indices

ub["candidate_wid_idx_list"] = ub["candidate_wid_list"].apply(map_wids_to_idx)
```

Candidate Encoder module
- For every candidate product (wid_idx), it:
  -  Looks up which brand, shop, and category levels it belongs to
  - Converts each of those categorical IDs into a dense vector (embedding)
  - Combines (sums) them into a single 64-dimensional candidate embedding vector
- Basically, given a product ID, give me a vector that summarizes all its attributes

```python
class CandidateEncoder(nn.Module):
    def __init__(self,
                 num_wid,
                 num_brand,
                 num_shop,
                 num_c1,
                 num_c2,
                 num_c3,
                 num_c4,
                 brand_lookup,
                 shop_lookup,
                 c1_lookup,
                 c2_lookup,
                 c3_lookup,
                 c4_lookup,
                 emb_dim=64):
        super().__init__()
        self.emb_dim = emb_dim

        # Embedding tables (0 index = padding/UNK)
        self.wid_emb   = nn.Embedding(num_wid,   emb_dim, padding_idx=0) # matrix of shape (num_wid, emb_dim = 64)
        self.brand_emb = nn.Embedding(num_brand, emb_dim, padding_idx=0)
        self.shop_emb  = nn.Embedding(num_shop,  emb_dim, padding_idx=0)
        self.c1_emb    = nn.Embedding(num_c1,    emb_dim, padding_idx=0)
        self.c2_emb    = nn.Embedding(num_c2,    emb_dim, padding_idx=0)
        self.c3_emb    = nn.Embedding(num_c3,    emb_dim, padding_idx=0)
        self.c4_emb    = nn.Embedding(num_c4,    emb_dim, padding_idx=0)

        # Register lookups as buffers (move with model.to(device), not trainable)
        self.register_buffer("brand_lookup", brand_lookup)
        self.register_buffer("shop_lookup",  shop_lookup)
        self.register_buffer("c1_lookup",    c1_lookup)
        self.register_buffer("c2_lookup",    c2_lookup)
        self.register_buffer("c3_lookup",    c3_lookup)
        self.register_buffer("c4_lookup",    c4_lookup)

    def forward(self, candidate_wid_idx):
        """
        candidate_wid_idx: LongTensor (B, C)
        Each entry is wid_idx (0 allowed as PAD).
        Returns: candidate_emb: (B, C, emb_dim)
        """
        wid_idx = candidate_wid_idx  # (B = number of query sessions, C = number of candidates per query)

        # Look up attribute indices for each wid_idx
        brand_idx = self.brand_lookup[wid_idx]  # (B, C)
        shop_idx  = self.shop_lookup[wid_idx]
        c1_idx    = self.c1_lookup[wid_idx]
        c2_idx    = self.c2_lookup[wid_idx]
        c3_idx    = self.c3_lookup[wid_idx]
        c4_idx    = self.c4_lookup[wid_idx]

        # Embed each attribute: all -> (B, C, emb_dim)
        wid_e   = self.wid_emb(wid_idx)
        brand_e = self.brand_emb(brand_idx)
        shop_e  = self.shop_emb(shop_idx)
        c1_e    = self.c1_emb(c1_idx)
        c2_e    = self.c2_emb(c2_idx)
        c3_e    = self.c3_emb(c3_idx)
        c4_e    = self.c4_emb(c4_idx)

        # Sum to get final candidate embedding (element wise sum)
        cand_emb = wid_e + brand_e + shop_e + c1_e + c2_e + c3_e + c4_e

        return cand_emb
```

```python
# Instantiate it
emb_dim = 64  # TBC NEEDS TO BE AGREED

cand_encoder = CandidateEncoder(
    num_wid=num_wid,
    num_brand=num_brand,
    num_shop=num_shop,
    num_c1=num_c1,
    num_c2=num_c2,
    num_c3=num_c3,
    num_c4=num_c4,
    brand_lookup=brand_lookup,
    shop_lookup=shop_lookup,
    c1_lookup=c1_lookup,
    c2_lookup=c2_lookup,
    c3_lookup=c3_lookup,
    c4_lookup=c4_lookup,
    emb_dim=emb_dim,
)
```

Example use
- the forward expects a (B, C) tensor but for every user, they have candidate wid idx list of vraiable lengths. since max is 200, we pad everyone with 0 until C = 200

```python
from torch.nn.utils.rnn import pad_sequence

# Convert each candidate list to a tensor
candidate_tensors = [
    torch.tensor(lst, dtype=torch.long)
    for lst in ub["candidate_wid_idx_list"]
]

# Pad them to equal length (max length = 200)
candidate_padded = pad_sequence(candidate_tensors, batch_first=True, padding_value=0)

# candidate_padded shape = (173831, 200)
print(candidate_padded.shape)

# Move both model and data to device (GPU if available)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
cand_encoder = cand_encoder.to(device)
candidate_padded = candidate_padded.to(device)


# Forward pass: get candidate embeddings
with torch.no_grad():  # disable gradient calc for inference/embedding extraction
    candidate_embeddings = cand_encoder(candidate_padded)

# Get embedding for a wid
wid = 84294794
wid_idx = wid2idx.get(wid, 0)

with torch.no_grad():
    vector = cand_encoder.wid_emb.weight[wid_idx]  # (64,)

vector.shape
```
